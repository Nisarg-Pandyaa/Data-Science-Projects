```python
import json
```

```python
#Function To Load Data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("Data Dump.json")
```

```python
#Function To Display User's Connections And Liked Pages
def display_users(data):
    print("USERS AND THEIR CONNECTIONS (Friends) & ACTIVITY :\n")

    for user in data["users"]:
        ku_id = user["KU'id"]
        name = user["name"]

        friend_list = []                 # Finds Friend's Names
        for friend in user["friends"]:
            for u in data["users"]:
                if u["KU'id"] == friend:
                    friend_list.append(u["name"])

        liked_pages = []                       # list of liked pages
        for like_page in user["liked_pages"]:
            for page in data["pages"]:
                if page["id"] == like_page:
                    liked_pages.append(page["name"])

        str_friend = ", ".join(friend_list) if friend_list else "Nobody"
        str_page = ", ".join(liked_pages) if liked_pages else "-"

        print(f"KU-ID : {ku_id} ~ {name} is Friend with {str_friend} and liked pages on topics : {str_page}")




    print("\nPAGES INFORMATION :\n")
    for page in data['pages']:
        print(f"{page['id']} : {page['name']}")

display_users(data)
```
